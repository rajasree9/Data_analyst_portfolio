```python
import pandas as pd

file_path = "/content/drive/MyDrive/sales_cleaned.csv"  # change if inside folder
df = pd.read_csv(file_path, encoding='latin1')

df.head()
```

```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
```

```python
df.dtypes
```

```python
sales_by_state = df.groupby('STATE')['SALES'].sum().sort_values(ascending=False)
sales_by_state.head(10)
```

```python
df['order_month'] = df['ORDERDATE'].dt.month
df['order_year'] = df['ORDERDATE'].dt.year
```

```python
sales_by_month = df.groupby(['order_year','order_month'])['SALES'].sum()
sales_by_month
```

```python
import matplotlib.pyplot as plt

sales_by_state.head(10).plot(kind='bar')
plt.title("Top 10 States by Sales")
plt.xlabel("State")
plt.ylabel("Total Sales")
plt.show()
```

```python
avg_sales_by_state = df.groupby('STATE')['SALES'].mean()
avg_sales_by_state.sort_values(ascending=False).head(10)
```

```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
```

```python
df['order_year'] = df['ORDERDATE'].dt.year
df['order_month'] = df['ORDERDATE'].dt.month
df['order_month_name'] = df['ORDERDATE'].dt.month_name()
```

```python
sales_by_month = (
    df.groupby(['order_year','order_month'])['SALES']
      .sum()
      .reset_index()
      .sort_values(['order_year','order_month'])
)

sales_by_month.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(sales_by_month['order_month'], sales_by_month['SALES'])
plt.title("Sales Trend by Month")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

```python
sales_by_state = (
    df.groupby('STATE')['SALES']
      .sum()
      .sort_values(ascending=False)
      .reset_index()
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.bar(sales_by_state['STATE'][:10],
        sales_by_state['SALES'][:10])

plt.title("Top 10 States by Sales")
plt.xlabel("State")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
top_products = (
    df.groupby('PRODUCTLINE')['SALES']
      .sum()
      .sort_values(ascending=False)
      .reset_index()
)
```

```python
plt.figure(figsize=(10,6))

plt.bar(top_products['PRODUCTLINE'][:10],
         top_products['SALES'][:10])

plt.title("Top 10 Products by Sales")
plt.ylabel("Total Sales")
plt.xlabel("Product Name")

plt.gca().invert_xaxis()   # Highest at top
plt.show()
```

```python
sales_by_category = (
    df.groupby('PRODUCTLINE')['SALES']
      .sum()
      .sort_values(ascending=False)
      .reset_index()
)

sales_by_category
```

```python
plt.figure(figsize=(8,5))
sales_by_category.plot(kind='bar', x='PRODUCTLINE', y='SALES')
plt.title("Sales by Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
sales_by_year = (
    df.groupby('order_year')['SALES']
      .sum()
      .reset_index()
      .sort_values('order_year')
)

sales_by_year
```

```python
import matplotlib.pyplot as plt

sales_by_year.plot(kind ='bar',x='order_year',y='SALES')

plt.title("Sales Trend by Year")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
avg_sales_by_year = (
    df.groupby('order_year')['SALES']
      .mean()
      .reset_index()
      .sort_values('order_year')
)

avg_sales_by_year
```

```python
sales_by_year['YoY_Growth_%'] = (
    sales_by_year['SALES'].pct_change() * 100
)

sales_by_year
```

```python

plt.plot(
    sales_by_year['order_year'],
    sales_by_year['YoY_Growth_%']
)

plt.title("Year-over-Year Sales Growth (%)")
plt.xlabel("Year")
plt.ylabel("Growth %")
plt.xticks(rotation=45)


plt.show()
```
